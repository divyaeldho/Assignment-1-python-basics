```python
# 1. Simple ATM Calculator

balance = int(input("Enter your account balance: "))# Get user's account balance


withdraw_amount = int(input("Enter withdrawal amount: "))# Get withdrawal amount

# Check if sufficient balance is available
if withdraw_amount <= balance:
    balance -= withdraw_amount
    print("Withdrawal successful!")
    print(f"Updated balance: {balance}")
else:
    print("Error: Insufficient balance.")
```

```python
#2. Electricity Bill Generator

units = int(input("Enter the number of units consumed: "))# Get the number of units consumed

bill = 0 # Initialize total bill

# Calculate bill based on slabs
if units <= 100:
    bill = units * 5
elif units <= 200:
    bill = (100 * 5) + ((units - 100) * 7)
else:
    bill = (100 * 5) + (100 * 7) + ((units - 200) * 10)

# Display total bill
print(f"Total bill: ₹{bill}")
```

```python
#3. Temperature Converter

# Get input
user_input = input("Enter temperature (e.g., 37C or 98.6F): ").upper()

# Extract numeric value and unit

temp = float(user_input[:-1])  # everything except last character
unit = user_input[-1]          # last character (C or F)

    # Conversion
if unit == 'C':
    fahrenheit = (temp * 9/5) + 32
    print(f"{temp}°C is equal to {fahrenheit}°F")
elif unit == 'F':
    celsius = (temp - 32) * 5/9
    print(f"{temp}°F is equal to {celsius}°C")
else:
    print("Invalid unit! Please use C for Celsius or F for Fahrenheit.")
```

```python
#Mobile Recharge Validity

from datetime import datetime, timedelta

# Get recharge date and today's date from user
recharge_date = input("Enter recharge date (dd-mm-yyyy): ")
today_date = input("Enter today's date (dd-mm-yyyy): ")

# Convert strings to date objects
recharge_date = datetime.strptime(recharge_date, "%d-%m-%Y").date()
today_date = datetime.strptime(today_date, "%d-%m-%Y").date()

# Calculate expiry date (28 days from recharge)
expiry_date = recharge_date + timedelta(days=28)

# Check if plan is valid or expired
if today_date <= expiry_date:
    days_left = (expiry_date - today_date).days
    print(f"Plan is valid. {days_left} days remaining.")
else:
    print("Plan expired.")
```

```python
#5. Bus Seat Reservation System

# Initial passenger list
passengers = ["Eldho", "Jacob", "Anna", "Divya", "Hazel", "Helna", "Hayden",]

print("Original passenger list:", passengers)

# Cancellation
cancel_name = input("Enter the name to cancel: ")
if cancel_name in passengers:
    passengers.remove(cancel_name)  # removes and shifts automatically
    print(f"{cancel_name} has been removed.")
else:
    print(f"{cancel_name} not found in the list.")

print("After cancellation:", passengers)

# New booking
new_name = input("Enter the new passenger name: ")
position = int(input(f"Enter position (1 to {len(passengers)+1}): "))

# Insert at given position
if 1 <= position <= len(passengers) + 1:
    passengers.insert(position - 1, new_name)
    print(f"{new_name} has been added at position {position}.")
else:
    print("Invalid position!")

print("Final passenger list:", passengers)
```

```python
#6. Sum of even numbers

# Get user input
n = int(input("Enter a number: "))

# Initialize sum
sum_of_even = 0

# Loop through numbers and add evens
for i in range(2, n + 1, 2):  # start from 2, step by 2 for evens
    sum_of_even += i

# Display result
print(f"Sum of all even numbers from 1 to {n} is {sum_of_even}")
```

```python
#7. Dictionary – Student Marks Summary

# Create an empty dictionary
students = {}

# Input data for 3 students
for i in range(1, 4):
    name = input(f"Enter name of student {i}: ")
    marks = float(input(f"Enter marks of {name}: "))
    students[name] = marks

# Calculate average marks
average_marks = sum(students.values()) / len(students)

# Display results
print( students)
print(f"Average Marks: {average_marks:}")
```

```python
#8. Dictionary – Word Frequency
# Get sentence from user
sentence = input("Enter a sentence: ")

# Defining function for word frequency counter
def word_frequency_counter(sentence):
    words = sentence.split()
    frequency = {}
    for word in words:
        if word in frequency:
            frequency[word] += 1
        else:
            frequency[word] = 1
    return frequency

#calling function
frequency = word_frequency_counter(sentence)

print("\nWord Frequencies:", frequency)
```

```python
#9. Functions – Check Prime

# define function for prime checker
def check_prime(number):
  for i in range(2,int(number/2)+1):
    if number%i==0:
      return False
  return True
# get input from user
number = int(input("Enter a number"))

#calling function
result = check_prime(number)

if result:
  print("The number is prime")
else:
  print("The number is not prime")
```

```python
#10. Vowel Counter

# defining function for vowel count
def count_vowels(text):
    vowels = "aeiouAEIOU"
    count = 0
    for char in text:
        if char in vowels:
            count += 1
    return count

# Get input from user and count vowels
string = input("Enter a string: ")
print(f"Number of vowels: {count_vowels(string)}")
```

```python
#11. Count Divisible Numbers

# Get inputs from user
N = int(input("Enter the range limit (N): "))
D = int(input("Enter the divisor (D): "))

# Count numbers divisible by D
count = 0
for i in range(1, N + 1):
    if i % D == 0:
        count += 1

# Display result
print(f"Numbers between 1 and {N} divisible by {D}: {count}")
```
